18◑ Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

Answer: For this assignment we used the text 3 (The Book of Genesis) 

```python
from collections import defaultdict
import operator
# Create bigram fro text3
bigramsWords1=nltk.bigrams(text3)
# Omitting the words from stopwords
bigramsWords2=[(k,y) for (k,y) in bigramsWords1 if not k.lower() in stoopwords and not y.lower() in stoopwords]
# Now create a dic to keep word and their occurance
bigramDic=defaultdict(int)
for a in bigramsWords2:
    bigramDic[a] += 1
bigramDicSorted = sorted(bigramDic.items(), key=operator.itemgetter(1))
bigramDicSorted.reverse()
bigramDicSorted[0:50]
```
